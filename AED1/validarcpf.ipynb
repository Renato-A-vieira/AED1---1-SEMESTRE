```python
from datetime import datetime

def validar_idade(idade):
    hoje = datetime.now().date()
    dia = int(idade[0:2])
    mes = int(idade[3:5])
    ano = int(idade[6:10])
    data_nasc = datetime(ano, mes, dia).date()
    idade = hoje.year - data_nasc.year - ((hoje.month, hoje.day) < (data_nasc.month, data_nasc.day))

    print(f"Sua idade é: {idade}")

    if idade >= 18:
        print("Você é maior de idade, pode cadastrar")
        return True
    else:
        print("Você é menor de idade, não pode cancelar")
        return False

def checar_cpf(digito):
    peso = len(digito) + 1
    soma = 0
    for ch in digito:
        soma += int(ch) * peso
        peso -= 1
    resto = soma % 11
    return 0 if resto < 2 else 11 - resto

def validar_cpf(cpf):
    if not cpf.isdigit() and len(cpf) != 11:
        print("CPF invalido, digite os 11 numeros")
        return False
    if cpf == cpf[0] * 11:
        print("CPF invalido")

    base = cpf[:9]

    d1 = checar_cpf(base)
    d2 = checar_cpf(base + str(d1))
    calc_cpf = base + str(d1) + str(d2)

    if cpf == calc_cpf:
        print("Cpf valido")
        return True

    else:
        print("CPF invalido")
        return False
    

def cadastrar():
    while True:
        nome = input("Digite seu Nome")
        sobrenome = input ("Digite seu Sobrenome")
        while True:
            cpf = input("Digite seu CPF ex: 11111111111")
            if validar_cpf(cpf):
                break
            else:
                print("Cpf Invalido")
                continue
        data_nasc = input("Digite sua data de nascimento ex: 11/11/1111")
        idade = data_nasc
        if not validar_idade(idade):
            print("Cadastro proibido")
            break

        renda = float(input("Digite sua renda ex: 1500.00"))

        print("\nSeu cadastro está completo")
        print(f" O {nome} {sobrenome} está apto")
        print(f" Seu cpf está valido :{cpf}")
        print(f" Idade : {idade}")
        print(f" Renda bruta: {renda}")
```

```python
def painel():

    print("\n" + "=" *40)
    print(f"{'Menu de cadastro':^40}")
    print("=" * 40)
    print("1 - Cadastrar")    
    print("2 - Exibir frase")
    print("3 - Sair")

    opcao = int(input("Digite a opção desejada"))
    if opcao == 1:
        cadastrar()
        break
    

painel()
```
