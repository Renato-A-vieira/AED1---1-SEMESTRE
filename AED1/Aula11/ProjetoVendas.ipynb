```python
texto = "Vendas de Agosto"

print(texto.center(30))

vendas_agosto = {
    "04/08/2025": 2350,
    "05/08/2025": 1980,
    "06/08/2025": 2540,
    "07/08/2025": 2100,
    "08/08/2025": 3050,
    "11/08/2025": 2650,
    "12/08/2025": 2180,
    "13/08/2025": 2760,
    "14/08/2025": 2920,
    "15/08/2025": 3100,
    "18/08/2025": 2300,
    "19/08/2025": 2430,
    "20/08/2025": 2210,
    "21/08/2025": 2790,
    "22/08/2025": 2970,
    "25/08/2025": 3100,
    "26/08/2025": 2880,
    "27/08/2025": 2600,
    "28/08/2025": 2500,
    "29/08/2025": 3400
}

total_vendas = list(vendas_agosto.values())
lista_dias = list(vendas_agosto.keys())

vendas_max = sum(total_vendas)

print(f"\033[1mTotal de vendas do mês de agosto foi:\033[0m {vendas_max}")

media_venda = vendas_max / len(total_vendas)

print(f"\033[1mA media de vendas do mês de agosto foi\033[0m: {media_venda}")

semana_1 = {}
semana_2 = {}
semana_3 = {}
semana_4 = {}

for i in range(0,5):
    semana_1[lista_dias[i]] = total_vendas[i]
    semana_2[lista_dias[i+5]] = total_vendas[i+5]
    semana_3[lista_dias[i+10]] = total_vendas[i+10]
    semana_4[lista_dias[i+15]] = total_vendas[i+15]

print("\n\033[1m Total Semanal \033[0m")
total_semana_1 = sum(list(semana_1.values()))
total_semana_2 = sum(list(semana_2.values()))
total_semana_3 = sum(list(semana_3.values()))
total_semana_4 = sum(list(semana_4.values()))

print(f"\n\033[1mA Total da primeira semana é: \033[0m {total_semana_1}")
print(f"\033[1mA Total da segunda semana é: \033[0m {total_semana_2}")
print(f"\033[1mA Total da terceira semana é: \033[0m {total_semana_3}")
print(f"\033[1mA Total da quarta semana é: \033[0m {total_semana_4}")

print("\n\033[1m Media Semanal \033[0m")

media_semana_1 = total_semana_1 / len(semana_1)
media_semana_2 = total_semana_2 / len(semana_2)
media_semana_3 = total_semana_3 / len(semana_3)
media_semana_4 = total_semana_4 / len(semana_4)

print(f"\n\033[1mA media da primeira semana é\033[0m : {media_semana_1}")
print(f"\033[1mA media da segunda semana é\033[0m : {media_semana_2}")
print(f"\033[1mA media da terceira semana é\033[0m : {media_semana_3}")
print(f"\033[1mA media da quarta semana é\033[0m : {media_semana_4}")

media_mes = vendas_max / 20


maior_mes = max(total_vendas)
indice_maior = total_vendas.index(maior_mes)
maior_venda = lista_dias[indice_maior]

print(f"\n\033[1m A data da maior venda foi:\033[0m {maior_venda} \033[1m e o valor foi:\033[0m {maior_mes}")

menor_mes = min(total_vendas)
indice_menor = total_vendas.index(menor_mes)
menor_venda = lista_dias[indice_menor]

print(f"\n\033[1m A data da menor venda foi: \033[0m {menor_venda} \033[1m e o valor foi:\033[0m {menor_mes}")

diferencial = ((maior_mes - media_mes) / media_mes*100)

print(f"\n\033[1m O Pico percentual é:\033[0m : {diferencial:.2f}%")
```
