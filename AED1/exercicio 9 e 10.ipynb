```python
frase = (input("Digite uma frase aqui"))
palavras = frase.lower().split()
contagem = {}
for palavra in palavras:
    if palavra in contagem:
        contagem[palavra] +=1
    else:
        contagem[palavra] = 1

print(contagem)
```

```python
vendedores = {
    "Carlos",
    "Lucas",
    "Amanda",
    "Renato",
    "João",
    "Celia",
    "Kleber",
    "Juliana" }

venda_vendedor = {
    "Carlos" : 2000,
    "Amanda" : 1500,
    "João" : 3000,
    "Renato" : 1000,
    "Renato" : 2100,
    "Amanda" : 900 }

vendedor_venda = list(venda_vendedor.keys())
vendas_total = list(venda_vendedor.values())

soma = sum(vendas_total)
media = sum(vendas_total) / len(vendas_total)

venda_maior = max(vendas_total)
indice_maior = vendas_total.index(venda_maior)
vendedor_maior = vendedor_venda[indice_maior]

com_vendas = set(venda_vendedor.keys())
sem_vendas = vendedores - com_vendas

print("\n\033[1m O valor total das vendas é \033[0m : ", soma)
print("\033[1m A Media de vendas é \033[0m : ", media)
print("\n\033[1m O vendedor", vendedor_maior,"vendeu:\033[0m" , venda_maior)
print(f"\033[1m Os vendedores que não venderam foram\033[0m", sem_vendas)
```

```python
vetor = []

for i in range (5):
    numero = int(input("Digite 5 numero inteiros"))
    vetor.append(numero)

maior_numero = max(vetor)
menor_numero = min(vetor)
pos_maior = vetor.index(maior_numero)
pos_menor = vetor.index(menor_numero)

print("\n\033[1m Os numero digitados foram:\033[0m", vetor)
print(f"\033[1m O maior numero digitado foi:\033[0m {maior_numero}. \033[1mEle está na posição:\033[0m {pos_maior}")
print(f"\033[1m O menor numero digitado foi:\033[0m {menor_numero}. \033[1mEle está na posição:\033[0m {pos_menor}")
                
```

```python
vetor_1 = []
vetor_2 = []
print("Digite os numeros para o 1º vetor")
for i in range(5):
    while True:
        try:
            numero_1 = int(input(f"Digite o {i+1}º numero para o primeiro vetor:"))
            vetor_1.append(numero_1)
            break
        except ValueError:
            print("Você não digitou ou digitou uma letra, digite novamente")

print("\nDigite os numeros para o 2º vetor")
for i in range(5):
    while True:
        try:
            numero_2 = int(input(f"Digite o {i+1}º numero para o segundo vetor:"))
            vetor_2.append(numero_2)
            break
        except ValueError:
            print("Você não digitou ou digitou uma letra, digite novamente")
            
intercalado = []

for i in range(5):
    intercalado.append(vetor_1[i])
    intercalado.append(vetor_2[i])
    

print (f"\n\033[1mLista do vetor 1:\033[0m {vetor_1}")
print (f"\033[1mLista do vetor 2:\033[0m {vetor_2}")
print (f"\033[1mLista intercalada :\033[0m {intercalado}")

    
```

```python
vetor = []

for i in range(20):
    while True:
        palavras = input(f"Digite a {i+1} palvavra (Max 10 caracteres)")
        if len(palavras) <= 10:
            vetor.append(palavras)
            break
        else:
                print("Palavra muito grande, digite novamente")
palavras_invertida = [palavra[::-1] for palavra in vetor]

print(palavras)
print(palavras_invertida)
```
