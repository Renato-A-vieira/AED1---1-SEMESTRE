```python
def gerar_imposto_renda (receita, despesas, imposto= 0.15, mostrar_resumo = True):
    lucro_bruto = receita - despesas
    if lucro_bruto == 0:
        print("Est√° zerado, gastou e recebeu o mesmo valor")
    elif lucro_bruto > 0:
        print("Est√° indo bem")
    else:
        lucro_bruto <= 0
        print("Est√° no prejuizo, est√° negativado")

    imposto_aplicado = lucro_bruto * imposto
    lucro_liquido = lucro_bruto - imposto_aplicado

    if mostrar_resumo:
        print("\nRelatorio Financeiro")
        print(f"\nReceita total: R$ {receita:.2f}")
        print(f"despesas total: R$ {despesas:.2f}")
        print(f"lucro bruto: R$ {lucro_bruto:.2f}")
        print(f"Imposto Aplicado: R$ {imposto_aplicado:.2f}")
        print(f"lucro liquido: R$ {lucro_liquido:.2f}")


gerar_imposto_renda (1000000, 1000)
```

```python
from random import choice

print("\33[1mJogo de papel, pedra tesoura\33[1m")

nome = str(input("Digite seu nome"))
op = ["pedra" , "papel" , "tesoura"]

vitoria_jogador = 0
vitoria_computador = 0
empate = 0

while True:
    print("\nEscolha: pedra, papel ou tesoura")
    jogador = input(f"{nome}, sua jogada: ").lower()

    computador = choice(op)
    print (f"O computador jogou : {computador}")

    if computador == jogador:
        print("Empate")
        empate +=1

    elif (jogador == "pedra" and computador == "tesoura") or \
         (jogador == "tesoura" and computador == "papel") or \
         (jogador == "papel" and computador == "pedra"):
         print(f" O jogador {nome}, venceu")
         vitoria_jogador +=1
    else:
        print("O computador venceu")
        vitoria_computador +=1
    continuar = input("Deseja jogar novamente? Digite sim ou n√£o")
    if continuar != 'sim':
        break
    
print("\n=== Resultado Final ===")
print(f"{nome}: {vitorias_jogador} vit√≥ria(s)")
print(f"Computador: {vitorias_computador} vit√≥ria(s)")
print(f"Empates: {empates}")

if vitorias_jogador > vitorias_computador:
    print(f"Parab√©ns, {nome}! Voc√™ foi o grande vencedor!")
elif vitorias_computador > vitorias_jogador:
    print("O computador ganhou! Tente novamente da pr√≥xima vez.")
else:
    print("ü§ù Empate geral!")
```
