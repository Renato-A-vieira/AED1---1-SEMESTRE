```python
from datetime import datetime

usuario = []

def validar_idade(data_nasc):
    hoje = datetime.now().date()
    dia = int(data_nasc[0:2])
    mes = int(data_nasc[3:5])
    ano = int(data_nasc[6:10])
    data_nasc = datetime(ano, mes, dia).date()
    idade = hoje.year - data_nasc.year - ((hoje.month, hoje.day) < (data_nasc.month, data_nasc.day))

    print(f"Sua idade é: {idade}")

    if idade >= 18:
        print("Você é maior de idade, pode cadastrar")
        return True
    else:
        print("Você é menor de idade, não pode cancelar")
        return False

def calcular_idade(data_nasc):
    hoje = datetime.now().date()
    dia = int(data_nasc[0:2])
    mes = int(data_nasc[3:5])
    ano = int(data_nasc[6:10])
    idade = hoje.year - ano - ((hoje.month, hoje.day) < (mes, dia))

    return idade


def checar_cpf(digito):
    peso = len(digito) + 1
    soma = 0
    for ch in digito:
        soma += int(ch) * peso
        peso -= 1
    resto = soma % 11
    return 0 if resto < 2 else 11 - resto

def validar_cpf(cpf):
    if not cpf.isdigit() or len(cpf) != 11:
        print("CPF invalido, digite os 11 numeros")
        return False
    if cpf == cpf[0] * 11:
        print("CPF invalido")
        return False
        
    base = cpf[:9]

    d1 = checar_cpf(base)
    d2 = checar_cpf(base + str(d1))
    calc_cpf = base + str(d1) + str(d2)

    if cpf == calc_cpf:
        print("Cpf valido")
        return True

    else:
        print("CPF invalido")
        return False
    

def cadastrar():
    while True:
        nome = input("Digite seu Nome")
        sobrenome = input ("Digite seu Sobrenome")
        while True:
            cpf = input("Digite seu CPF <ex: 11111111111> :")
            if validar_cpf(cpf):
                break
            else:
                print("Cpf Invalido")
                continue
        data_nasc = input("Digite sua data de nascimento ex: 11/11/1111")
        if not validar_idade(data_nasc):
            print("Cadastro proibido")
            continue

        idade = calcular_idade(data_nasc)       
       
        renda = float(input("Digite sua renda -- ex 1500.00:"))

        cadastro = {
            "nome": nome,
            "sobrenome": sobrenome,
            "cpf": cpf,
            "idade": idade,
            "renda": renda
        }

        usuario.append(cadastro)
        
        
        print('\n' + "=" * 40)
        print("Seu cadastro está completo")
        print("=" * 40)
        print('\n')
        print(f" O usuario \033[1m{nome} {sobrenome}\033[0m está apto")
        print(f" Seu cpf está valido :{cpf}")
        print(f" Idade: {idade}")
        print(f" Renda bruta: {renda}")
        break
```

```python
def painel():
    while True: 
        print("\n" + "=" *40)
        print(f"{'Menu de cadastro':^40}")
        print("=" * 40)
        print("1 - Cadastrar")    
        print("2 - Exibir frase")
        print("3 - Sair")
    
        opcao = int(input("Digite a opção desejada"))
        if opcao == 1:
           cadastrar()
        elif opcao == 2:
            print("\033[1m Seus sonhos não precisam de plateia, eles só precisam de você! \033[0m")
        elif opcao == 3:
            print("Obrigado e até a proxima")
            break
        else:
            print("fim")
        
    
painel()
```
