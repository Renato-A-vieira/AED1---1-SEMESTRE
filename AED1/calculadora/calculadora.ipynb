```python
from tkinter import *
from tkinter import ttk

#cores
cor1= "#f7f4f2"
cor2= "#121111"
cor3= "#06080d"
cor4= "#b56262"
cor5= "#525252"

janela =Tk()
janela.title("Calculadora")
janela.geometry("241x300")
janela.config(bg=cor1)

#frames}
frame_tela = Frame(janela, width=245, height=40, bg=cor1,)
frame_tela.grid(row=0, column=0)

frame_corpo = Frame(janela, width=245, height=310, bg=cor5)
frame_corpo.grid(row=1, column=0)

#função
todos_valores = ''
valor_texto = StringVar()
def calcular(valortotal):

    global todos_valores    
    todos_valores = todos_valores + str(valortotal)
    
    valor_texto.set(todos_valores)

def resultado():
    global todos_valores
    total = eval(todos_valores)
    valor_texto.set(str(total))

def limpar():
    global todos_valores    
    todos_valores = ""
    valor_texto.set("")
#label

app_label = Label(frame_tela, textvariable=valor_texto , bg=cor1, fg=cor2, anchor ="e", justify=RIGHT, font=('ivy', 20, 'bold'))
app_label.place(x=0, y=0, width=245, height=40)

#botões
b_1 = Button(frame_corpo, command = limpar, text="C", width =10, height=2, bg=cor4, fg=cor1,  font=('ivy', 12, 'bold'))
b_1.place(x=0, y=0)

b_2 = Button(frame_corpo, command = lambda : calcular('/100'),  text="%", width =6, height=2, bg=cor4, fg=cor1,  font=('ivy', 12, 'bold'))
b_2.place(x=110, y=0)

b_3 = Button(frame_corpo, command = lambda : calcular('/'), text="/", width =6, height=2, bg=cor4, fg=cor1,  font=('ivy', 12, 'bold'))
b_3.place(x=180, y=0)

b_4 = Button(frame_corpo, command = lambda : calcular('7'), text="7", width =5, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_4.place(x=0, y=52)

b_5 = Button(frame_corpo, command = lambda : calcular('8'), text="8", width =5, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_5.place(x=61, y=52)

b_6 = Button(frame_corpo, command = lambda : calcular('9'), text="9", width =5, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_6.place(x=122, y=52)

b_7 = Button(frame_corpo, command = lambda : calcular('*'), text="*", width =5, height=2, bg=cor4, fg=cor1,  font=('ivy', 12, 'bold'))
b_7.place(x=183, y=52)

b_8 = Button(frame_corpo, command = lambda : calcular('4'), text="4", width =5, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_8.place(x=0, y=104)

b_9 = Button(frame_corpo, command = lambda : calcular('5'), text="5", width =5, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_9.place(x=61, y=104)

b_10 = Button(frame_corpo, command = lambda : calcular('6'), text="6", width =5, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_10.place(x=122, y=104)

b_11 = Button(frame_corpo, command = lambda : calcular('-'),  text="-", width =5, height=2, bg=cor4, fg=cor1,  font=('ivy', 12, 'bold'))
b_11.place(x=183, y=104)

b_12 = Button(frame_corpo, command = lambda : calcular('1'), text="1", width =5, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_12.place(x=0, y=156)

b_13 = Button(frame_corpo, command = lambda : calcular('2'), text="2", width =5, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_13.place(x=61, y=156)

b_14 = Button(frame_corpo, command = lambda : calcular('3'), text="3", width =5, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_14.place(x=122, y=156)

b_15 = Button(frame_corpo, command = lambda : calcular('+'), text="+", width =5, height=2, bg=cor4, fg=cor1,  font=('ivy', 12, 'bold'))
b_15.place(x=183, y=156)

b_1 = Button(frame_corpo, command = lambda : calcular('0'),  text="0", width =10, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_1.place(x=0, y=208)

b_2 = Button(frame_corpo, command = lambda : calcular('.'), text=".", width =6, height=2, bg=cor2, fg=cor1,  font=('ivy', 12, 'bold'))
b_2.place(x=110, y=208)

b_3 = Button(frame_corpo, command = resultado,  text="=", width =6, height=2, bg=cor4, fg=cor1,  font=('ivy', 12, 'bold'))
b_3.place(x=180, y=208)


janela.mainloop()
```
